```python
def splitData(corpus,train_ratio,valid_ratio,test_ratio):
    with open(corpus, 'r') as f:
        text = f.readlines()
    text = [line.strip() for line in text if line.strip()]
    train_size = int(len(text) * train_ratio)
    valid_size = int(len(text) * valid_ratio)
    test_size = int(len(text) * test_ratio)
    train_data = text[:train_size]
    valid_data = text[train_size:train_size + valid_size]
    test_data = text[train_size + valid_size:]
    return train_data, valid_data, test_data

train_data, valid_data, test_data = splitData(filePath,0.7,0.15,0.15)
trainset = Dataset(train_data, 32)

```

```python
import torch
from torch import nn

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"Using device {device}")

# use validation data set also to check the model performance
class LSTMmodel(nn.Module):
    def __init__(self, embedding_size, hidden_size, num_layers, epsilon, momentum, trainset,valset):
        super(LSTMmodel, self).__init__()
        self.trainset = trainset
        self.vocab_size = len(trainset.vocab)
        self.embedding_size = embedding_size
        self.hidden_size = hidden_size
        self.num_layers = num_layers
        self.epsilon = epsilon
        self.momentum = momentum
        self.validationset = valset
        self.embedding = nn.Embedding(
            self.vocab_size, embedding_size, device=device)
        self.lstm = nn.LSTM(input_size=self.embedding_size, hidden_size=self.hidden_size,
                            num_layers=self.num_layers, batch_first=True, device=device)
        self.output = nn.Linear(
            self.hidden_size, self.vocab_size, bias=False, device=device)

    def forward(self, xContext):
        xembed = self.embedding(xContext)
        hidden, (h0, c0) = self.lstm(xembed)
        output = self.output(hidden[:, -1])
        return output
    
    def validation_step(self):
        valData = [line for line in self.validationset if line.strip()]
        with torch.no_grad():
            total_loss = 0
            total_correct = 0
            total_samples = 0
            for line in valData:
                tokens = Tokenise(line)
                # tokens = [w for w in tokens if w in self.trainset.vocab]
                ind = [self.trainset.wordToIndex.get(
                    w, self.trainset.unKnownIndex) for w in tokens]
                sent = [ind[:i] for i in range(1, len(ind))]
                sent = [([self.trainset.startIndex]+[self.trainset.padIndex] *
                        (self.trainset.max_len - len(prf)) + prf) for prf in sent]
                sent = torch.tensor(sent, dtype=torch.long, device=device)
                target = torch.tensor(
                    ind[1:], dtype=torch.long, device=device)
                output = self.forward(sent)
                loss = nn.CrossEntropyLoss()(output, target)
                total_loss += loss.item()
                total_samples += 1
                total_correct += (output.argmax(1) == target).sum().item()
        return total_loss / total_samples, total_correct / total_samples
                    
        
            


```
